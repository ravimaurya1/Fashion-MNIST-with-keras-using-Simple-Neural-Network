```python
class_labels=['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']
```

## Preprocess the Dataset

```python
plt.figure()
plt.imshow(train_images[3],cmap=plt.cm.binary)
plt.colorbar()
plt.show()
```

## Normalize Dataset

```python
train_images1=train_images/255.0
test_images1=test_images/255.0
```

## Show 25 images of our Dataset

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_labels[train_labels[i]])
plt.show()
```

## Build the Neural Network

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

## Convert target set to categorical vector

```python
from tensorflow.keras.utils import to_categorical

train_labels = to_categorical(train_labels)
test_labels  = to_categorical(test_labels)
```

## Compile the model

```python
model.compile(optimizer='adam', 
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

## Train the model

```python
model.fit(train_images1,train_labels,epochs=5)
```

So here Accuracy of model is 88.38% on our training set

## Evaluate accuracy on testing set

```python
test_loss, test_acc = model.evaluate(test_images1, test_labels)

print('Test accuracy:', test_acc)
```

## Now make prediction

```python
predictions = model.predict(np.expand_dims(test_images1[9], axis=0))
plt.imshow(test_images[9],cmap=plt.cm.binary)
print ("Prediction=",class_labels[np.argmax(predictions)],np.max(predictions)*100,"%")
print ("Actual=",class_labels[np.argmax(test_labels[9])])
```

## Lets predict for 10 images

```python
num_rows = 10
num_cols = 3
num_images = num_rows*num_cols
plt.figure(figsize=(2*num_cols, 2*num_rows))
plt.grid(False)
plt.axis('off')
for i in range(num_images):
    plt.subplot(num_rows, num_cols, i+1)
    plt.xticks([])
    plt.yticks([])
    predictions = model.predict(np.expand_dims(test_images1[i], axis=0))
    plt.imshow(test_images[i],cmap=plt.cm.binary)
    xd="Prediction="+str(class_labels[np.argmax(predictions)])+'\n'+str(round(np.max(predictions)*100,2))#+"%"+"Actual="+str(class_labels[np.argmax(test_labels[i])])
    plt.xlabel(xd)
```
